# Basic audio generation

```python
# generate a sine wave of 220 Hz and 440 Hz and the sum of them

import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavfile
import os

# sampling rate
fs = 44100

sin_220 = np.sin(2*np.pi*220*np.arange(10*fs)/fs)
sin_440 = np.sin(2*np.pi*440*np.arange(10*fs)/fs)
sin_sum = sin_220 + sin_440

# plot the first 100 samples
plt.plot(sin_220[:100])
plt.plot(sin_440[:100])
plt.plot(sin_sum[:100])
plt.show()

# write the audio files
base_path = "noise_and_edge_cases"

wavfile.write(os.path.join(base_path, 'sin_220.wav'), fs, sin_220)
wavfile.write(os.path.join(base_path, 'sin_440.wav'), fs, sin_440)
wavfile.write(os.path.join(base_path, 'sin_sum.wav'), fs, sin_sum)
```

```python
# random noise

noise = np.random.normal(0, 1, 10*fs)

# plot the first 100 samples
plt.plot(noise[:100])

# write the audio file
wavfile.write(os.path.join(base_path, 'noise.wav'), fs, noise)
```

```python
# convert to mp3, ogg, and webm

path = "speech/es_male_speech_and_noises.wav"

# convert to mp3
os.system("ffmpeg -i " + path + " " + path[:-4] + ".mp3")

# convert to ogg
os.system("ffmpeg -i " + path + " " + path[:-4] + ".ogg")

# convert to webm
os.system("ffmpeg -i " + path + " " + path[:-4] + ".webm")
```
